# Import libraries

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold, cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from pprint import pprint
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
```

# Upload/import the file


```
# upload the file into google colab
from google.colab import files
uploaded = files.upload()
```

```
# import the file into a dataframe
import io
df = pd.read_csv(io.BytesIO(uploaded['ML_4colour.csv']))
```

# Data cleaning

```
# convert labels to numerical values
print(df['hull'].value_counts())
df['hull1'] = df['hull'].astype('category')
df.dtypes
df['hull1'] = df['hull1'].cat.codes
```

```
# drop the rows with missing data
df = df.drop(df[(df.cat == 0)| (df['1525.1'] == 0) ].index)
print(len(df))
```

# Original Model

```
x = np.asarray(df.drop(['Folder','genotypes','location','hull','cotyledon','rep',
                         '4_sal','4_coum','3_cinn','sal','cat_13C','hull1'], axis = 1))

y= np.asarray(df['hull1'])
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, test_size = 0.2, random_state = 100)
print(len(x_train))
print(len(x_test))
model = RandomForestClassifier()
model.fit(x_train, y_train)
prediction_test = model.predict(x_test)
prediction_train = model.predict(x_train)
accuracy_train = metrics.accuracy_score(y_train, prediction_train)
accuracy_test = metrics.accuracy_score(y_test, prediction_test)
print(accuracy_train, accuracy_test)
# 1.0 0.6578947368421053
kf = StratifiedKFold(n_splits=10)
scores = cross_val_score(model, x_train, y_train, cv=kf)
print(pd.DataFrame(scores), np.mean(scores), np.std(scores))
#0.6241758241758243 0.1667043742313796
```

# Improving the original model by decreasing the features

```
# Finding important features
df1 = df.drop(['Folder','genotypes','location','hull','cotyledon','rep',
                         '4_sal','4_coum','3_cinn','sal','cat_13C','hull1'], axis = 1)
x_train = pd.DataFrame(x_train, columns = df1.columns)
feature_importances = pd.DataFrame(model.feature_importances_,
                                   index = x_train.columns,
                                    columns=['importance']).sort_values('importance',
                                                                        ascending=False)
feature_importances = feature_importances.round(2)
print(feature_importances.head(10))
```

```
# making new model based on more important features
x = np.asarray(df[['cat_gluc','gallo','epi_cat','cat','595.6I','1492.2I']])
y= np.asarray(df['hull1'])
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, test_size = 0.2, random_state = 100)
print(len(x_train))
print(len(x_test))
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
prediction_test = model2.predict(x_test)
prediction_train = model2.predict(x_train)
accuracy_train = metrics.accuracy_score(y_train, prediction_train)
accuracy_test = metrics.accuracy_score(y_test, prediction_test)
print(accuracy_train, accuracy_test)
# 0.993421052631579 0.7894736842105263
kf = StratifiedKFold(n_splits=10)
scores2 = cross_val_score(model2, x_train, y_train, cv=kf)
print(pd.DataFrame(scores), np.mean(scores2), np.std(scores2))
#0.7563873626373627 0.12788824858214126
```

# Hyperparameters tuning (Random search)

```
# Look at parameters used by our current forest
pprint(model2.get_params())
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node (internal node that have children)
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node (external node without children)
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)


# Use the random grid to search for best hyperparameters
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
model2_random = RandomizedSearchCV(estimator = model2, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
model2_random.fit(x_train, y_train)
model2_random.best_params_
#{'n_estimators': 452,
 'min_samples_split': 10,
 'min_samples_leaf': 2,
 'max_features': 'sqrt',
 'max_depth': 60,
 'bootstrap': True}
```

# Hyperparameters tuning (Grid search)

```
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [40,60,80],
    'max_features': [2, 3,4],
    'min_samples_leaf': [2,3, 4],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [350, 450, 550, 650]
}
# Create a based model
grid_search = GridSearchCV(estimator = model2, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
# Fit the grid search to the data
grid_search.fit(x_train, y_train)
grid_search.best_params_
#{'bootstrap': True,
 'max_depth': 40,
 'max_features': 4,
 'min_samples_leaf': 2,
 'min_samples_split': 8,
 'n_estimators': 350}
```

# Evaluation

```
def evaluate(model, x_test, y_test):
    model.fit(x_train, y_train)
    prediction_test = model.predict(x_test)
    prediction_train = model.predict(x_train)
    accuracy_train = metrics.accuracy_score(y_train, prediction_train)
    accuracy_test = metrics.accuracy_score(y_test, prediction_test)
    print(accuracy_train, accuracy_test)
    return accuracy_train, accuracy_test


model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
model2 = evaluate(model2, x_test, y_test)

model2_random = model2_random.best_estimator_
random_accuracy = evaluate(model2_random, x_test, y_test)

grid_search = grid_search.best_estimator_
grid_accuracy = evaluate(grid_search, x_test, y_test)

#0.9802631578947368 0.7368421052631579
#0.9802631578947368 0.7368421052631579
#0.875 0.8157894736842105 # great improvement
```

# The best model

```
# apply the tuned hyperparameters to the model2, check the accuracy, and visualize
# keeping the important features that with available chemical standards
x = np.asarray(df[['cat_gluc','gallo','epi_cat','cat','lut_4gluc','myr_3rhamn']])
y= np.asarray(df['hull1'])
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, test_size = 0.2, random_state = 100)
best_model = RandomForestClassifier(bootstrap=True, max_depth=40, max_features=4,min_samples_leaf=2,
                       min_samples_split=8,n_estimators=350, random_state=150)
best_model.fit(x_train, y_train)
prediction_test = best_model.predict(x_test)
prediction_train = best_model.predict(x_train)
accuracy_train = metrics.accuracy_score(y_train, prediction_train)
accuracy_test = metrics.accuracy_score(y_test, prediction_test)
print(accuracy_train,accuracy_test)
# 0.868421052631579 0.8421052631578947
```
